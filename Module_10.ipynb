## Nd Array

```python
import numpy as np

#1d array

arr1=np.array([2,4,1,2,5])

arr1=(arr1*5)

print(arr1)

print(arr1.ndim)

# 2d array 
arr2 = np.array([ [1,2,3], # row1
                  [4,5,6] #row2
                ])

#3d array 

arr3 = np.array([
    #1st floor
    [ [1,2,3], # row1
      [4,5,6]  ], #row2 

    #2nd floor

     [[1,2,3], # row1
      [4,5,6]  ], #row2  


    #3rd floor
      [[1,2,3], # row1
      [4,5,6]  ] #row2  

    
])
```

### Array attributes

```python
#1.demention
print(arr1.ndim)
print(arr2.ndim)
print(arr3.ndim)

#2.shape

print(arr1.shape)
print(arr2.shape)
print(arr3.shape)

#3.Data_type

print(arr1.dtype)
print(arr2.dtype)
print(arr3.dtype)

#4 size

print(arr1.size)
print(arr2.size)
print(arr3.size)
```

### ndarray datatype

```python
arr = np.array([1,2,3]) 

print(arr.dtype)

arr = np.array([1.5,2.4,3.2]) 
print(arr.dtype)
```

```python
#Upcasting 
arr = np.array([1,2,3.2])
print(arr.dtype)

arr = np.array([1,2,3.2,"hello"])
print(arr.dtype)

arr = np.array([1,True,3.25,'hello',]) 
print(arr.dtype)
```

```python
#selecting datatype on array


arr = np.array([1,2,30],dtype=np.int8)

print(arr.dtype)

arr=arr.astype(np.int32)

print(arr.dtype)

#error
arr = np.array([1,2,300],dtype=np.int8)

arr = np.array([1,2,3.2,'hello'],dtype=np.float64) 

print(arr.dtype)
```

## indexing and slicing

```python
#indexing

print(arr1)
# print(arr2)

arr1[2]=100
print(arr1)

print(arr1[2])


# print(arr1)
arr2[0][2]= 100

print(arr2)
```

```python
#slicing 
#1d = arr[start:end:step] 

# arr_mod=arr1[1:4]

# arr_mod[1]=200

# print(arr1)

# print(arr_mod)

arr_mod=arr1[1:4].copy()

arr_mod[2]=100
print(arr_mod)
print(arr1)
```

```python
#2d slicing 
# arr[ row_start : row_end : step , column_start : column_end : step ]

print(arr2) 

getting_row=arr2[0:1:,]
getting_row1=arr2[1::,]

print(getting_row)
print(getting_row1)



#getting a column 

col_0 = arr2[:: ,0:1]
col_1 = arr2[::,1:2]

print(col_0)
print(col_1)


#getting a portion 


portion = arr2[:: , 0:2] 

print(portion)
```
