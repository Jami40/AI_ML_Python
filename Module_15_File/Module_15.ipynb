## Bar Chart

```python
data = {
    'Student': [ 'B', 'C', 'D', 'E', 'F', 'G', 'H'],
    'Gender': [ 'Male', 'Female', 'Male', 'Female', 'Male', 'Female', 'Male'],
    'StudyHours': [ 2, 5, 3, 7, 4, 8, 1],
    'Attendance': [ 60, 88, 70, 95, 80, 98, 55],
    'Grade': [ 'Fail', 'Pass', 'Fail', 'Pass', 'Pass', 'Pass', 'Fail']
}

df = pd.DataFrame(data) 
```

```python
df
```

```python
gender_group=df.groupby('Gender').size()

gender_group

gender_group.index
gender_group.values
```

```python
plt.bar(gender_group.index,gender_group.values,color=['orange','red'],edgecolor='blue')

plt.xlabel('Gender')
plt.ylabel('Count')
plt.title('Bar Chart of Gender among Students')
```

```python
# Bar chart on real data_

df=pd.read_csv('student_data (1).csv')

df
```

```python
## Bar Chart on Completion Status
status=df.groupby('CompletionStatus').size()

plt.bar(status.index,status.values,color=['orange','red','blue'],edgecolor='black')

plt.xlabel('Completion Satatus')
plt.ylabel('Count')
plt.title('Bar Chart of Completion Status among Students')

```

```python
## Bar Chart on Location 
status=df.groupby('Location').size()

plt.bar(status.index,status.values,edgecolor='black')

plt.xlabel('Loaction')
plt.ylabel('Count')
plt.title('Bar Chart of Loaction among Students')

plt.show() ### Its has great functionality 
```

## Pie Chart

```python
status = df.groupby('CompletionStatus').size()
plt.pie(status,labels=status.index,autopct='%1.1f%%',explode = (0.1,0,0),shadow=True)
plt.show()
```
